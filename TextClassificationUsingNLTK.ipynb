```python
import numpy as np
import pandas as pd
import codecs
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
import nltk
```

```python
#save stop words
def getStopWords() :
    import string
    stop = stopwords.words('english')
    punctuations = list(string.punctuation)
    stop = stop + punctuations
    return stop
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag) :
    if tag.startswith('J') :
        return wordnet.ADJ
    elif tag.startswith('V') :
        return wordnet.VERB
    elif tag.startswith('N') :
        return wordnet.NOUN
    elif tag.startswith('R') :
        return wordnet.ADV
    else :
        return wordnet.NOUN
```

```python
def cleanWords(words) :
    stop_words = getStopWords()
    lemmatizer = WordNetLemmatizer()
    clean = []
    for w in words :
        w2 = w.lower()
        if(len(w2) < 3 or w2 in stop_words) :
            continue
        pos = pos_tag([w])
        cw = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
        cw = cw.lower()
        clean.append(cw)
    return clean
```

```python
# get all features(words) from a file
def readfile(path) :
    with codecs.open(path, "r",encoding='utf-8', errors='ignore') as f:
        content =f.read()
    words = word_tokenize(content)
#     for r in replacements:
#         content = content.replace(r, ' ')
#     words = content.strip().split(" ")
    clean_words = cleanWords(words)
    return clean_words
```

```python
def getfreq(words) :
    dictionary = {}
    for w in words :
        if(w in dictionary) :
            dictionary[w] = dictionary[w] + 1
        else :
            dictionary[w] = 1
    return dictionary
```

```python
def getRow(features,words) :
    row = np.zeros(len(features))
    frequency = getfreq(words)
    j = 0
    for f in features :
        if f in frequency :
            row[j] = frequency[f]
        j = j + 1
    return row
                
```

```python
def fit(X,Y) :
    result = {}
    classes = set(Y)
    result["total_documents"] = X.shape[0]
    for current_class in classes :
        result[current_class] = {}
        number_rows = (Y == current_class)
        X_current = X[number_rows]
        Y_current = Y[number_rows]
        all_words = X.shape[1]
        result[current_class]["docs_with_current_class"] = len(X_current)
        result[current_class]["total_words"] = X_current.sum()
        for j in range(all_words) :
            result[current_class][j] = X_current[ : , j].sum()
    return result
```

```python
def probability(x,dictionary,current_class) :
    prob = np.log(dictionary[current_class]["docs_with_current_class"]) - np.log(dictionary["total_documents"])
    for j in range(len(x)) :
            if x[j] != 0 :
                prob += np.log(dictionary[current_class][j] + 1) - np.log(dictionary[current_class]["total_words"] + len(x))
    return prob
```
